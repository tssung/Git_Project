```python
import requests

# 取得台灣銀行牌告匯率
res = requests.get('http://rate.bot.com.tw/xrt?Lang=zh-TW')
print (res.text)
```

```python
# 使用 pandas 取得台灣銀行牌告匯率
import pandas

dfs = pandas.read_html('http://rate.bot.com.tw/xrt?Lang=zh-TW')
```

```python
type(dfs)
```

```python
len(dfs)
```

```python
# 顯示取得的內容
dfs[0]
```

```python
currency=dfs[0]
```

```python
type(currency)
```

```python
# currency.ix[:,0:5] 已不被使用,用以下取代
currency.iloc[:,0:5]
```

```python
# 取得前五個欄位資料
currency = currency.iloc[:,0:5]
```

```python
currency
```

```python
# 設定 column name
# 中文設定 unicode
currency.columns=[u'幣別',u'現金匯率-本行買入',u'現金匯率-本行賣出',u'即期匯率-本行買入',u'即期匯率-本行賣出']
```

```python
currency
```

```python
# 取得幣別 USD,HKD,...
currency[u'幣別'] = currency[u'幣別'].str.extract('\((\w+)\)')
```

```python
currency
```

```python
# 將 currency 資料存到 excel file
currency.to_excel('currency.xlsx')
```
